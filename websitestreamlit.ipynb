```python
# Install Streamlit in Colab
!pip install streamlit

# Create the Streamlit Python script
import os

# Create a Python file with the Streamlit code
with open("app.py", "w") as f:
    f.write("""
import streamlit as st

# Set up the title of the web app
st.title("My First Streamlit Web App")

# Add a header
st.header("This is a simple web app built with Streamlit")

# Display a simple message
st.write("Streamlit is a great tool to quickly create interactive web applications with Python.")

# Add some widgets for interaction
name = st.text_input("Enter your name:")
if name:
    st.write(f"Hello, {name}! Welcome to the Streamlit app.")

# Add a button and display a message when clicked
if st.button('Say Hello'):
    st.write('Hello there!')

# Display a slider
slider_value = st.slider("Select a number:", 0, 100, 25)
st.write(f"Slider value: {slider_value}")

# Display a plot (Example: Simple Line Chart)
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
st.pyplot(plt)
""")

# Run Streamlit from within Colab
!streamlit run app.py & npx localtunnel --port 8501
```
